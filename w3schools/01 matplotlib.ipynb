## Matplotlib Labels

<b>Create Labels for a Plot</b>

With Pyplot, you can use the xlabel () and ylabel() functions to seet a label for the x- and y-axis.

```python
import sys
import matplotlib

# matplotlib.use('Agg') # 作用是在PyCharm中不顯示繪圖

import numpy as np
import matplotlib.pyplot as plt

x = np.array([80, 85, 90, 95, 100, 105, 110, 115, 120, 125])
y = np.array([240, 250, 260, 270, 280, 290, 300, 310, 320, 330])

font1 = {'family':'serif','color':'blue','size':20}
font2 = {'family':'serif','color':'darkred','size':15}

plt.title("Sports Watch Data", fontdict = font1)
plt.xlabel("Average Pulse", fontdict = font2)
plt.ylabel("Calorie Burnage", fontdict = font2)

plt.plot(x, y)
plt.show()
```

### Display Multiple Plots

With the subplot() function you can draw multiple plots in one figure(圖表)

```python
import matplotlib.pyplot as plt
import numpy as np

# plot 1
x = np.array([0, 1, 2, 3])
y = np.array([3, 8, 1, 10])

# the figure has 1 row, 2 columns, and this plot is the first plot.
plt.subplot(1, 3, 1) # 一行，三列中的第一列
plt.plot(x,y)

# plot 2 
x = np.array([0, 1, 2, 3])
y = np.array([10, 20, 30, 40])

plt.subplot(1, 3, 2) # 三個圖表中的第二個
plt.plot(x,y)

plt.subplot(1,3, 3) # 三個圖表中的第三個
plt.plot(x,y)

plt.show()
```

### Matplotlib Scatter（散點圖）

<b>Creating Scatter Plots</b>

With Pyplot,you can use the scatter() function to draw a scatter plot.

The scatter() function plots one dot for each observation. It needs two arrays of the same length. one for the values of the x-axis,and one for values on the y-axis.

```python
# A simple scatter plot:

import matplotlib.pyplot as plt
import numpy as np

x = np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
y = np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])

plt.scatter(x, y)
plt.show()

print('''
The observation in the example above is the result of 13 cars passing by.
The X-axis shows how old the car is.
The Y-axis shows the speed of the car when it passes.

=============================================================================
Q：Are there any relationships between the observations?
A：It seems that the newer the car, the faster it drives, but that could be a coincidence, after all we only registered 13 cars.
''')
```

<b>Compare Plots（比較圖）</b>

In teh example above, there seems to be a relationship between speed and age, but what if we plot the observations from another day as well? Will the scatter plot tell us something else?

在上面的例子中，速度和年齡之間似乎存在關係，但是如果也繪製另一天的觀察結果呢?散點圖會告訴我們其他信息嗎？

```python
# Draw two plots on the same figure:

import matplotlib.pyplot as plt
import numpy as np

# day one, the age and speed of 13 cars:
x = np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
y = np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])
plt.scatter(x, y)

# day two, the age and speed of 15 cars:
x = np.array([2,2,8,1,15,8,12,9,7,3,11,4,7,14,12])
y = np.array([100,105,84,105,90,99,90,95,94,100,79,112,91,80,85])
plt.scatter(x, y)

plt.show()

print(
'''
Note：The two plots are plotted with two different colors, by default blue and orange.

==========================================================================================
By comparing the two plots, it is safe to say that they both gibes us the same conclusion: the newer the car, the faster it drives.
'''
)
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
y = np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])
colors = np.array(["red","green","blue","yellow","pink","black","orange","purple","beige","brown","gray","cyan","magenta"])

plt.scatter(x, y, c=colors)

plt.show()
```

<b>Color Map</b>

The Matplotlib module has a number of available colormaps.You can include the colormap in the drawing by including the plt.colorbar() statement.

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
y = np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])
colors = np.array([0, 10, 20, 30, 40, 45, 50, 55, 60, 70, 80, 90, 100])

# You can specify the colormap with the keyword argument 'camp' with the value of the colormap
plt.scatter(x, y, c=colors, cmap='viridis')

plt.colorbar()

plt.show()
```

<b>Size</b>

Change the size of the dots with the s argument.

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
y = np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])
sizes = np.array([20,50,100,200,500,1000,60,90,10,300,600,800,75])


# Alpha
# adjust the transparency of the dots with the alpha argument
plt.scatter(x, y, s=sizes, alpha=0.5)

plt.show()
```

<b>Combine Color Size and Alpha</b>

Combine a colormap with different size on the dots. This is best visualized if the dots are transparent.

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.random.randint(100, size=(100))
y = np.random.randint(100, size=(100))
colors = np.random.randint(100, size=(100))
sizes = 10 * np.random.randint(100, size=(100))

plt.scatter(x, y, c=colors, s=sizes, alpha=0.5, cmap='nipy_spectral')

plt.colorbar()

plt.show()
```

### Creating Bars

With Pyplot, you can use the bar() function to draw bar grarphs.

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.array(["A", "B", "C", "D"])
y = np.array([3, 8, 1, 10])

plt.bar(x, y)
plt.show()
```

<b>Horizontal Bars</b>

To be displayed horizontally instead of vertically, use the barh() function.

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.array(["A", "B", "C", "D"])
y = np.array([3, 8, 1, 10])

plt.barh(x, y)
plt.show()
```

<b>Bar Color</b>

The bar() and barh() takes the keyword argument color to set the color of the bars.

<b>Bar Width</b>

The bar() takes the keyword argument width to set the width of the bars.

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.array(["A", "B", "C", "D"])
y = np.array([3, 8, 1, 10])

plt.bar(x, y, color = "red", width = 0.1)
# plt.barh(x, y, height = 0.1)
plt.show()
```

### Creating Pie Charts

Add labels to the pie chart with the label paraeter.

```python
import matplotlib.pyplot as plt
import numpy as np

y = np.array([35, 25, 25, 15]) # 餅圖是逆時針開始的
mylabels = ["Apples", "Bananas", "Cherries", "Dates"]

plt.pie(y, labels = mylabels)
plt.show() 
```

<b>Start Angle</b>

As mentioned the defaulte star angle is at the x-axis, but youcan change the start angle by specifying a startangle parameter.

如前所述，默認起始角度位於x軸，但可以通過指定 startangle 參數來更改起始角度。

```python
import matplotlib.pyplot as plt
import numpy as np

y = np.array([35, 25, 25, 15])
mylabels = ["Apples", "Bananas", "Cherries", "Dates"]

plt.pie(y, labels = mylabels, startangle = 90)
plt.show() 
```

<b>Explode</b>

The explode parameter allows you to let one of the wedges to stand out.

```python
import matplotlib.pyplot as plt
import numpy as np

y = np.array([35, 25, 25, 15])
mylabels = ["Apples", "Bananas", "Cherries", "Dates"]
myexplode = [0.2, 0, 0, 0]

# Add a shadow to the pie chart by setting the shadows parameter to True.
plt.pie(y, labels = mylabels, explode = myexplode, shadow = True)
plt.show() 
```

<b>Colors</b>

Set the color of each wedge with the colors parameter.

```python
import matplotlib.pyplot as plt
import numpy as np

y = np.array([35, 25, 25, 15])
mylabels = ["Apples", "Bananas", "Cherries", "Dates"]
mycolors = ["white", "hotpink", "b", "#4CAF50"]

plt.pie(y, labels = mylabels, colors = mycolors)
plt.show() 
```

<b>Legend</b>

To add a list of explanation for each wedge, use the legend() function.

```python
import matplotlib.pyplot as plt
import numpy as np

y = np.array([35, 25, 25, 15])
mylabels = ["Apples", "Bananas", "Cherries", "Dates"]

plt.pie(y, labels = mylabels)
plt.legend(title = "Four Fruits:")
plt.show() 
```
